## Insertamos datos tabla Oscars

```python
#leemos el csv

df_oscars = pd.read_csv("CSV_JSON_F4/lista_tuplas_fase4.csv", index_col = 0)

#Hacemos una lista con el nombre de las columnas actuales del data frame
columnas_actuales = df_oscars.columns.tolist()

#Definimos el nombre de las columnas
nuevos_nombres = ['FechaCeremonia', 'MejorPelicula', 'MejorDirector', 'MejorActor', 'MejorActriz']

#Cambiamos el nombre de las columnas por la lista que hemos creado
df_oscars.columns = nuevos_nombres
```

```python

# Conectar a la base de datos
cnx = mysql.connector.connect(user='root', password='MySQLPassword2023!',
                              host='127.0.0.1', database='moviemakers')
mycursor = cnx.cursor()



#Iteramos por cada uno de los elementos

for index, row in df_oscars.iterrows():
    sql = """
        INSERT INTO Oscars (FechaCeremonia, MejorPelicula, MejorDirector, MejorActor, MejorActriz) 
        VALUES ( %s, %s, %s, %s, %s)
    """
    values = (row['FechaCeremonia'], row['MejorPelicula'], row['MejorDirector'], row['MejorActor'], row['MejorActriz'])
    
    try:
        mycursor.execute(sql, values)
    except mysql.connector.Error as err:
        print(f"Error: {err}")
        print("Error Code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)

# Confirmar la inserción
cnx.commit()

print(f'Se han insertado los datos correctamente')
# Cerrar conexión
mycursor.close()
cnx.close()
```

## Insertamos datos tabla películas

```python
#leemos el csv

df_peliculas = pd.read_csv("CSV_JSON_F1/resultado_tuplas_total.csv")  #cambiar el nombre del archivo

#Hacemos una lista con el nombre de las columnas actuales del data frame
columnas_actuales = df_peliculas.columns.tolist()

#Definimos el nombre de las columnas
nuevos_nombres = ['Tipo', 'Nombre', 'AnioEstreno', 'MesEstreno', 'IdPelicula']

#Cambiamos el nombre de las columnas por la lista que hemos creado
df_peliculas.columns = nuevos_nombres



#Combertimos en INT las columnas AnioEstreno y MesEstreno porque daba error de NAN Además convertimos el ano y mes en INT

df_peliculas['AnioEstreno'] = df_peliculas['AnioEstreno'].fillna(0).astype(int)
df_peliculas['MesEstreno'] = df_peliculas['MesEstreno'].fillna(0).astype(int)

#Eliminamos duplicados 
df_peliculas = df_peliculas.drop_duplicates()

```

```python

# Conectar a la base de datos
cnx = mysql.connector.connect(user='root', password='MySQLPassword2023!',
                              host='127.0.0.1', database='moviemakers')
mycursor = cnx.cursor()



#Iteramos por cada uno de los elementos

for index, row in df_peliculas.iterrows():           
    sql = """
        INSERT INTO Peliculas (Tipo, Nombre, AnioEstreno, MesEstreno, IdPelicula) 
        VALUES ( %s, %s, %s, %s, %s)
    """
    values = (row['Tipo'], row['Nombre'], row['AnioEstreno'], row['MesEstreno'], row['IdPelicula'])
    
    try:
        mycursor.execute(sql, values)
    except mysql.connector.Error as err:
        print(f"Error: {err}")
        print("Error Code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)

# Confirmar la inserción
cnx.commit()

print(f'Se han insertado los datos correctamente')
# Cerrar conexión
mycursor.close()
cnx.close()
```

## Insertamos datos tabla caracteristicas películas

```python
#leemos el csv

df_caracteristicas_peliculas = pd.read_csv("CSV_JSON_F2/archivo_final_actualizado.csv")  #cambiar el nombre del archivo

#Hacemos una lista con el nombre de las columnas actuales del data frame
columnas_actuales = df_caracteristicas_peliculas.columns.tolist()

#Definimos el nombre de las columnas
nuevos_nombres = ['PuntuacionIMDB', 'PuntuacionTomatoes', 'Directores', 'Guionistas', 'Argumento', 'Genero', 'DuracionMinutos', 'Nombre', 'idPelicula']

#Cambiamos el nombre de las columnas por la lista que hemos creado
df_caracteristicas_peliculas.columns = nuevos_nombres

#Eliminamos duplicados 
df_caracteristicas_peliculas = df_caracteristicas_peliculas.drop_duplicates()

# cambiamos en el dataframe la , por . lo convertimos en float y quitomos los errorer Nan

df_caracteristicas_peliculas['PuntuacionIMDB'] = df_caracteristicas_peliculas['PuntuacionIMDB'].str.replace(',', '.')
df_caracteristicas_peliculas['PuntuacionIMDB'] = df_caracteristicas_peliculas['PuntuacionIMDB'].fillna('Null').astype(float)
df_caracteristicas_peliculas['PuntuacionTomatoes'] = df_caracteristicas_peliculas['PuntuacionTomatoes'].fillna('Null').astype(float)
df_caracteristicas_peliculas['Directores'] = df_caracteristicas_peliculas['Directores'].fillna('Null')
df_caracteristicas_peliculas['Guionistas'] = df_caracteristicas_peliculas['Guionistas'].fillna('Null')
df_caracteristicas_peliculas['Argumento'] = df_caracteristicas_peliculas['Argumento'].fillna('Null')
df_caracteristicas_peliculas['Genero'] = df_caracteristicas_peliculas['Genero'].fillna('Null')
df_caracteristicas_peliculas['DuracionMinutos'] = df_caracteristicas_peliculas['DuracionMinutos'].fillna('Null')
df_caracteristicas_peliculas['Nombre'] = df_caracteristicas_peliculas['Nombre'].fillna('Null')
```

```python
# Conectar a la base de datos
cnx = mysql.connector.connect(user='root', password='MySQLPassword2023!',
                              host='127.0.0.1', database='moviemakers')
mycursor = cnx.cursor()



#Iteramos por cada uno de los elementos

for index, row in df_caracteristicas_peliculas.iterrows():           
    sql = """
        INSERT INTO Peliculas_caracteristicas (PuntuacionIMDB, PuntuacionTomatoes, Directores, Guionistas, Argumento, Genero, DuracionMinutos, Nombre, idPelicula) 
        VALUES ( %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """
    values = (row['PuntuacionIMDB'], row['PuntuacionTomatoes'], row['Directores'], row['Guionistas'], row['Argumento'], row['Genero'], row['DuracionMinutos'] , row['Nombre'], row['idPelicula'])
    
    try:
        mycursor.execute(sql, values)
    except mysql.connector.Error as err:
        print(f"Error: {err}")
        print("Error Code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)

# Confirmar la inserción
cnx.commit()

print(f'Se han insertado los datos correctamente')
# Cerrar conexión
mycursor.close()
cnx.close()
```
